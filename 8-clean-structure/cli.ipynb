## ðŸŽ¯ What is `argparse`?

`argparse` is Pythonâ€™s built-in module for **parsing command-line arguments**.  
It helps you build CLI tools that accept:
- options like `--verbose`
- values like `--file data.csv`
- flags like `-v`

You define what arguments your script expects â€” and `argparse` does the parsing, error checking, and help text generation automatically.

---

## âœ… Why Use `argparse`?

| Without `argparse`         | With `argparse`                        |
|----------------------------|----------------------------------------|
| You manually parse `sys.argv` | âœ… Automatic parsing & validation     |
| No help or usage output    | âœ… `--help` works by default            |
| No input validation        | âœ… Type checking and default values     |
| Not user-friendly          | âœ… Clear, professional CLI experience   |

---

## ðŸš€ Basic Example

```python
# cli_tool.py
import argparse

parser = argparse.ArgumentParser(description="Simple calculator")
parser.add_argument("a", type=int, help="First number")
parser.add_argument("b", type=int, help="Second number")
parser.add_argument("--operation", choices=["add", "sub"], default="add")

args = parser.parse_args()

if args.operation == "add":
    result = args.a + args.b
else:
    result = args.a - args.b

print(f"Result: {result}")
```

### âœ… Usage:

```bash
python cli_tool.py 4 2 --operation sub
# Result: 2

python cli_tool.py 4 2
# Result: 6 (default is add)

python cli_tool.py --help
```

âœ… Output:
```
usage: cli_tool.py [-h] [--operation {add,sub}] a b
```

| Concept              | Example                                      |
|----------------------|----------------------------------------------|
| Positional argument  | `a`, `b` â€” must be provided                  |
| Optional argument    | `--operation` â€” optional with a default      |
| Type casting         | `type=int` â†’ converts input to integer       |
| Help text            | `help="..."` shows up in `--help`            |
| Choice restriction   | `choices=[...]` limits valid values          |

## ðŸ“¦ Organizing CLI Entry Point

Best practice: use a `main()` function and a guard:

```python
def main():
    parser = argparse.ArgumentParser()
    # define args
    args = parser.parse_args()
    # use args...

if __name__ == "__main__":
    main()
```

# âš¡ï¸ Building Command-Line Interfaces with `Typer`

## ðŸŽ¯ What is `Typer`?

**Typer** is a modern Python library for building **user-friendly and powerful CLI tools** â€” based on:
- Pythonâ€™s type hints
- FastAPI-style syntax
- `Click` under the hood (but simpler and cleaner)

### âœ¨ Why is Typer great?

| Feature              | Benefit                                       |
|----------------------|-----------------------------------------------|
| Type-safe            | Uses type hints to parse/validate inputs      |
| Automatic help       | `--help` docs generated automatically         |
| Multiple commands    | Supports subcommands with ease                |
| Developer-friendly   | You just write functions!                     |
| Fast to prototype    | Minimal code to go from idea to CLI           |

---

## ðŸ“¦ Installation

```bash
pip install typer[all]
```

> The `[all]` extra installs `rich` too, for pretty help and errors.

---

## ðŸš€ Basic Example

```python
# hello.py
import typer

app = typer.Typer()

@app.command()
def hello(name: str, uppercase: bool = False):
    """Say hello to someone."""
    greeting = f"Hello, {name}"
    if uppercase:
        greeting = greeting.upper()
    typer.echo(greeting)

if __name__ == "__main__":
    app()
```

---

### âœ… Usage

```bash
python hello.py Alice
# Hello, Alice

python hello.py Alice --uppercase
# HELLO, ALICE

python hello.py --help
```

âœ… Output:
```
Usage: hello.py [OPTIONS] NAME
```

---

## ðŸŽ¯ Key Concepts

| Feature           | Example                         |
|-------------------|----------------------------------|
| Positional arg    | `name: str`                     |
| Optional flag     | `uppercase: bool = False`       |
| Help text         | Docstring of the function       |
| Type conversion   | Automatic from type hints       |
| Output            | `typer.echo()` for safe printing|

---

## ðŸ” Multiple Commands

```python
app = typer.Typer()

@app.command()
def add(x: int, y: int):
    """Add two numbers."""
    typer.echo(x + y)

@app.command()
def subtract(x: int, y: int):
    """Subtract y from x."""
    typer.echo(x - y)
```

```bash
python cli.py add 2 3
python cli.py subtract 10 4
```

---

## ðŸ§  Advanced Features

- `typer.Option()` â†’ for setting help, default, validation
- `typer.Argument()` â†’ for customizing positional args
- Call other commands with `app.invoke(...)`
- Nest subcommands via `app = typer.Typer()` and `main_app.add_typer(app)`

---

## ðŸ§ª Example with File and Option

```python
@app.command()
def read(file: typer.FileText, limit: int = typer.Option(5, help="How many lines to read")):
    for i, line in enumerate(file, 1):
        if i > limit:
            break
        typer.echo(f"{i}: {line.strip()}")
```

```bash
python cli.py read data.txt --limit 3
```
