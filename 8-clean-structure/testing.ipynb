## ðŸ§ª What is `unittest`?

`unittest` is the **built-in testing framework** in Python.  
It follows the **xUnit** style (used in Javaâ€™s JUnit, .NETâ€™s NUnit, etc.).

You define tests by **subclassing** `unittest.TestCase` and writing methods that begin with `test_`.

---

## ðŸ§± Basic Structure of a `unittest` Test Case

```python
import unittest

def add(a, b):
    return a + b

class TestMath(unittest.TestCase):
    def test_add_positive(self):
        self.assertEqual(add(2, 3), 5)

    def test_add_negative(self):
        self.assertEqual(add(-1, -2), -3)

if __name__ == "__main__":
    unittest.main()
```

---

## ðŸ“š Common `assertX` Methods

| Method                        | Meaning                                 |
|------------------------------|-----------------------------------------|
| `assertEqual(a, b)`          | a == b                                  |
| `assertNotEqual(a, b)`       | a != b                                  |
| `assertTrue(x)`              | bool(x) is True                         |
| `assertFalse(x)`             | bool(x) is False                        |
| `assertIs(a, b)`             | a is b                                  |
| `assertIsNone(x)`            | x is None                               |
| `assertIn(a, b)`             | a in b                                  |
| `assertRaises(Exception)`    | Checks if error is raised               |

## âš™ï¸ Setup and Teardown

Use `setUp()` and `tearDown()` to run logic before and after every test:

```python
class MyTests(unittest.TestCase):
    def setUp(self):
        self.data = [1, 2, 3]

    def tearDown(self):
        print("Cleaning up!")

    def test_sum(self):
        self.assertEqual(sum(self.data), 6)
```

## ðŸš€ Running Tests

You can run the tests in different ways:

- From the command line:
```bash
python -m unittest test_module.py
```

- Or with test discovery:
```bash
python -m unittest discover
```

By default, Python will discover tests in files that match:  
ðŸ“„ `test*.py`

## ðŸ§ª Testing Exceptions

```python
def divide(a, b):
    return a / b

class TestDivide(unittest.TestCase):
    def test_zero_division(self):
        with self.assertRaises(ZeroDivisionError):
            divide(1, 0)
```

## ðŸ§ª Mocking (Bonus!)

To test external dependencies like APIs or databases:

```python
from unittest.mock import patch

@patch("my_module.get_data_from_api")
def test_api(mocked_api):
    mocked_api.return_value = {"key": "value"}
    result = my_module.process_data()
    assert result == ...
```

# ðŸš€ Getting Started with `pytest`

## ðŸŽ¯ Why Use `pytest`?

- Is **more readable**
- Supports **advanced features** like fixtures, parameterization, and plugins

> âœ… Write less code, test more things, get better feedback.

## ðŸ“¦ Installation

```bash
pip install pytest
```

---

## ðŸ§± Basic Structure

Just write functions starting with `test_`, and use regular `assert` statements.

### ðŸ§ª Example:

```python
# math_utils.py
def add(a, b):
    return a + b

# test_math_utils.py
from math_utils import add

def test_add_positive():
    assert add(2, 3) == 5

def test_add_negative():
    assert add(-1, -2) == -3
```

Then run with:

```bash
pytest
```

---

## âœ… Features Youâ€™ll Love

### ðŸ”¹ 1. No need to subclass `TestCase`
Just write plain functions â€” much simpler than `unittest`.

---

### ðŸ”¹ 2. Native `assert` is powerful

You can write:

```python
assert add(2, 2) == 4
```

And if it fails, `pytest` shows **detailed error introspection**:

```
>       assert add(2, 2) == 5
E       assert 4 == 5
```

---

### ðŸ”¹ 3. Parametrized Tests (ðŸ“Œ Powerful!)

Write **one test**, run it with many inputs:

```python
import pytest
from math_utils import add

@pytest.mark.parametrize("a,b,expected", [
    (1, 2, 3),
    (-1, -1, -2),
    (100, 200, 300)
])
def test_add(a, b, expected):
    assert add(a, b) == expected
```

---

### ðŸ”¹ 4. Fixtures â€“ Setup & Teardown

Use fixtures for reusable setup logic:

```python
import pytest

@pytest.fixture
def sample_data():
    return {"a": 1, "b": 2}

def test_data_keys(sample_data):
    assert "a" in sample_data
```

> Fixtures are **more flexible** than `setUp()` in `unittest`.

---

### ðŸ”¹ 5. Auto Test Discovery

Just run:

```bash
pytest
```

Pytest will automatically:
- Find files like `test_*.py`
- Run functions that start with `test_`

## ðŸ§ª Testing Exceptions

```python
import pytest

def divide(a, b):
    return a / b

def test_zero_division():
    with pytest.raises(ZeroDivisionError):
        divide(1, 0)
```
