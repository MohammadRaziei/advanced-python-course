## üß∞ OpenCV vs PIL: A Practical Comparison

### ‚úÖ OpenCV: Computer Vision Powerhouse

**OpenCV (cv2)** is an open-source library optimized for real-time computer vision.

Let‚Äôs add **Salt & Pepper noise** to an image and remove it with **median filtering**:



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'assets/panda.jpg'
original_image = cv2.imread(image_path)

# Add salt & pepper noise
def add_salt_pepper_noise(image, prob):
    noisy = np.copy(image)
    num_noisy = int(prob * image.size)
    coords = [np.random.randint(0, i, num_noisy // 2) for i in image.shape]
    noisy[tuple(coords)] = 255  # Salt
    coords = [np.random.randint(0, i, num_noisy // 2) for i in image.shape]
    noisy[tuple(coords)] = 0    # Pepper
    return noisy

noisy_image = add_salt_pepper_noise(original_image, 0.4)
filtered_image = cv2.medianBlur(noisy_image, 5)

# Show result
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1); plt.title("Original"); plt.imshow(original_image); plt.axis('off')
plt.subplot(1, 3, 2); plt.title("Noisy"); plt.imshow(noisy_image); plt.axis('off')
plt.subplot(1, 3, 3); plt.title("Filtered"); plt.imshow(filtered_image); plt.axis('off')
plt.tight_layout()
plt.show()
```

‚úÖ Pros:
- Very fast
- Low-level control
- Ideal for computer vision tasks

‚ùó Note: OpenCV loads images in **BGR** format (not RGB) by default.

---

### üñºÔ∏è PIL / Pillow: Simple and Pythonic

**Pillow** (modern fork of PIL) is great for image manipulation and basic editing.

```python
from PIL import Image
import matplotlib.pyplot as plt

original_image = Image.open('assets/panda.jpg')
resized = original_image.resize((300, 300))
rotated = original_image.rotate(45, expand=True)
grayscale = original_image.convert('L')

plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1); plt.title("Original"); plt.imshow(original_image); plt.axis('off')
plt.subplot(2, 2, 2); plt.title("Resized"); plt.imshow(resized); plt.axis('off')
plt.subplot(2, 2, 3); plt.title("Rotated"); plt.imshow(rotated); plt.axis('off')
plt.subplot(2, 2, 4); plt.title("Grayscale"); plt.imshow(grayscale, cmap='gray'); plt.axis('off')
plt.tight_layout()
plt.show()
```

‚úÖ Pros:
- Very Pythonic
- Easy-to-use
- Lightweight
